```python
rep = {'ã':'a','á':'a','é':'e','í':'i','ó':'o','ô':'o','ú':'u','`a':'a','è':'e','ì':'i','ò':'o','ù':'u','ç':'c','ñ':'n','š':'s','ø':''}

datos['forename'] =datos['forename'].replace(rep, regex=True)
```

```python
import unicodedata 
def remove_accents(input_str):
    nkfd_form = unicodedata.normalize('NFKD', input_str) 
    only_ascii = nkfd_form.encode('ASCII', 'ignore') 
    only_ascii
    
    return only_ascii
```

```python
# Apply to your df
datos_decode= datos["forename"].apply(remove_accents)
```

```python

datos["forename"]=datos_decode.str.decode('utf8', errors='strict')
```
